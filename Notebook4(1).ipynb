# Movimiento planetario
Esta dado por la ley de gravitacion universal

  $$m \frac{d^2\vec r}{dt²}= -GM_s m_p\frac{\vec r}{|r|^3}$$
  
  El movimiento no depende de la masa del planeta,$m_p$, podemos hacer el cambio de variable:
  $T=\sqrt{\tfrac{GM_s}{R_T^3}}t$, $\vec R=\tfrac{\vec r}{R_T}$ y la ecuación anterior queda de la siguiente manera:
  
  $$\frac{d²\vec R}{dT²}= -\frac{\vec R}{|R|³}$$

$G= 6.67.4\times 10^{-11} \tfrac{N m^2}{kg^2}$

$M_s=1.9891\times 10^{30} kg$

$R_T=1.52\times 10^{11} m$

Librerias de Ipython que utilizaremos:

```python
from pylab import*
from scipy.integrate import*
%matplotlib inline
```

```python
G=6.674e-11
M_s=1.9891e30
R_T = 1.52e11
Cte =sqrt((G*M_s)/(R_T**3))
yr=60*60*24*365
P=(2*pi)/Cte
print P/yr
print P,yr
```

```python
def gravitacion(r_v,t):
    #r_v=(x,y,vx,vy)
    r=sqrt(r_v[0]**2+r_v[1]**2)
    dx=r_v[2]
    dy=r_v[3]
    dvx=-r_v[0]/r**3
    dvy=-r_v[1]/r**3
    return [dx,dy,dvx,dvy]
```

```python
tiempo=linspace(0,2*pi)
sol=odeint(gravitacion,(1,0,0,1),tiempo)
x,y=sol[:,0],sol[:,1]
xlim([-1.5,1.5]),ylim([-1.5,1.5]),axes().set_aspect('equal') #preferencias de la gráfica
plot(x,y)
plot(0,0,'o')
show()
```

```python
tiempo=linspace(0,2*pi)
vx=0.5
vy=sqrt(1-vx**2)
sol=odeint(gravitacion,(1,0,vx,vy),tiempo)
x,y=sol[:,0],sol[:,1]
xlim([-1.5,1.5]),ylim([-1.5,1.5]),axes().set_aspect('equal') #preferencias de la gráfica
plot(x,y,'o')
plot(0,0,'o')
show()
```

```python
tiempo=linspace(0,2*pi,500)
for num_trayectorias in range(10):
    vx=rand()
    vy=sqrt(1-vx**2)
    sol=odeint(gravitacion,(1,0,vx,vy),tiempo)
    x,y=sol[:,0],sol[:,1]
    plot (x,y)
xlim([-1.5,2.0]),ylim([-1.5,2.0]),axes().set_aspect('equal') #preferencias de la gráfica
plot(0,0,'o')
show()
```
